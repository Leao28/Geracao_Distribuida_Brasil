##  **2. Distribuição de Frequências**

```python
def Dist_Freq(Parametro):
    frequencia_absoluta = df_Base[Parametro].value_counts()
    frequencia_percentual = (df_Base[Parametro].value_counts(normalize = True) * 100).round(2)
    distribuicao = pd.DataFrame({'Frequencia': frequencia_absoluta, 'Percentual (%)': frequencia_percentual})
    distribuicao.rename_axis(Parametro, axis = 'columns', inplace = True)
    return distribuicao

def Dist_Freq_2(Parametro_1, Parametro_2):
    frequencia_absoluta = df_Base.groupby([Parametro_1])[Parametro_2].value_counts()
    frequencia_percentual = (df_Base.groupby([Parametro_1])[Parametro_2].value_counts(normalize = True) * 100).round(2)
    distribuicao = pd.DataFrame({'Frequencia': frequencia_absoluta, 'Percentual (%)': frequencia_percentual})
    return distribuicao
```

```python
Dist_Freq('Classe_Consumo')\
    .style.format({'Percentual (%)': '{:.2f}'})\
    .bar(subset = ['Percentual (%)'], color='lightgreen', height=50, width=60)
```

```python
Dist_Freq_2('Classe_Consumo', 'Regiao')\
    .style.format({'Percentual (%)': '{:.2f}'})\
    .bar(subset = ['Percentual (%)'], color='lightgreen', height=50, width=60)
```
